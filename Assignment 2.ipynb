# Question 2

```python
import nltk
stopword = nltk.corpus.stopwords.words('english')
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stopword
```

```python
lower = []
for i in corpus:
    lower.append(' '.join([word.lower() for word in i.split()]))

normalized_data = []
for i in lower:
    normalized_data.append(' '.join([ps.stem(word) for word in i.split() if word.isalpha() and word not in stopword]))
normalized_data

    
```

# Question 3

```python
filenames = []
for i in range(100):
    filenames.append('review' + str(i+1))
filenames
```

(a)

```python
bin_tdm_dict = {}
for i in range(len(filenames)):
    word_count = {}
    for word in normalized_data[i].split():
        if word not in word_count.keys():
            word_count[word] = 1
    bin_tdm_dict[filenames[i]] = word_count
bin_tdm_dict
```

```python
binary_tdm = pd.DataFrame(bin_tdm_dict)
binary_tdm.fillna(0, inplace = True)
binary_tdm
```

(b)

```python
termfreq_tdm_dict = {}
for i in range(len(filenames)):
    word_count = {}
    for word in normalized_data[i].split():
        if word not in word_count.keys():
            word_count[word] = 1
        else:
            word_count[word] += 1
    termfreq_tdm_dict[filenames[i]] = word_count
termfreq_tdm_dict
```

```python
termfreq_tdm = pd.DataFrame(termfreq_tdm_dict)
termfreq_tdm.fillna(0 ,inplace = True)
termfreq_tdm
```

(c)

```python
tdm_lenght_norm = pd.DataFrame(termfreq_tdm_dict)
tdm_lenght_norm.fillna(0, inplace = True)
tdm_lenght_norm = tdm_lenght_norm.iloc[:,:].div(tdm_lenght_norm.sum(axis=0),axis=1)
tdm_lenght_norm
```

(d)

```python
no_of_docs = len(filenames)
unique_terms = termfreq_tdm.index
idf_score_dict = {}
for term in unique_terms:
    idf_score_dict[term] = 0
    for i in normalized_data:
        if term in i.split():
            idf_score_dict[term] += 1
    idf_score_dict[term] = np.log10(no_of_docs/idf_score_dict[term])
idf_score_dict
```

```python
idf_score_series = pd.Series(idf_score_dict)
idf_score = pd.DataFrame(idf_score_series)
idf_score
```

```python
tf_idf_tdm = termfreq_tdm*idf_score.values
tf_idf_tdm
```
